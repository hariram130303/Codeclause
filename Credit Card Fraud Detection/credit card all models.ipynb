```python
#Data Visualization
def draw_histograms(dataframe, features, rows, cols):
    fig = plt.figure(figsize=(20, 20))
    for i, feature in enumerate(features):
        ax = fig.add_subplot(rows, cols, i+1)
        dataframe[feature].hist(bins=20, ax=ax, facecolor='midnightblue')
        ax.set_title(feature + " Distribution", color="DarkRed")
        ax.set_yscale("log")
    fig.tight_layout()
    plt.show()

draw_histograms(df, df.columns, 8, 4)
```

```python
def draw_boxplots(dataframe, features, rows, cols):
    fig = plt.figure(figsize=(20, 20))
    for i, feature in enumerate(features):
        ax = fig.add_subplot(rows, cols, i+1)
        dataframe.boxplot(column=feature, ax=ax, vert=False)
        ax.set_title(feature + " Distribution", color="DarkRed")
    fig.tight_layout()
    plt.show()

draw_boxplots(df, df.columns, 8, 4)
```

```python
#Feature Analysis
import matplotlib.pyplot as plt
import seaborn as sns

def plot_boxplots(dataframe, feature, target):
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=target, y=feature, data=dataframe)
    plt.title(f"{feature} Distribution - Non-Fraud vs Fraud", fontsize=16)
    plt.xlabel("Fraudulent Transaction", fontsize=12)
    plt.ylabel(feature, fontsize=12)
    plt.show()

# Example usage:
plot_boxplots(df, "Amount", "Class")
```

```python
def class_imbalance_analysis(dataframe, target):
    class_counts = dataframe[target].value_counts()
    positive_count = class_counts[1]
    negative_count = class_counts[0]
    imbalance_ratio = positive_count / negative_count

    print("Class Imbalance Analysis:")
    print("=========================")
    print("Positive (Fraudulent) Instances:", positive_count)
    print("Negative (Non-Fraudulent) Instances:", negative_count)
    print("Imbalance Ratio (Positive to Negative):", imbalance_ratio)

# Example usage:
class_imbalance_analysis(df, "Class")
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB,GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score,ConfusionMatrixDisplay,confusion_matrix,classification_report
```

# Splitting the variables

```python
X=df.drop('Class',axis=1)
y=df['Class']
```

# Train Test Split 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Models

```python
log=LogisticRegression(max_iter=2000,solver='saga')
log.fit(X_train,y_train)
```

```python
svc=SVC()
svc.fit(X_train,y_train)
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from collections import Counter
```

```python
importances = rfc.feature_importances_
indices = np.argsort(importances)[::-1]
selected_features_tree = X.columns[indices[:5]].tolist()
```

```python
X=df[selected_features_tree]
y=df.iloc[:,-1]
```

```python
# the numbers before smote
num_before = dict(Counter(y))

#perform smoting

# define pipeline
over = SMOTE(sampling_strategy=0.8)
under = RandomUnderSampler(sampling_strategy=0.8)
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)

# transform the dataset
X_smote, y_smote = pipeline.fit_resample(X, y)


#the numbers after smote
num_after =dict(Counter(y_smote))
```

```python
print(num_before, num_after)
```

```python
# visualizing the change
labels = ["Negative Cases","Positive Cases"]
plt.figure(figsize=(8,3))
plt.subplot(1,2,1)
sns.barplot(x=labels, y=(list(num_before.values())))
plt.title("Numbers Before Balancing")
plt.subplot(1,2,2)
sns.barplot(x=labels, y=(list(num_after.values())))
plt.title("Numbers After Balancing")
plt.show()
```

# Predicting

```python
a=[]
p=[]
r=[]
f=[]
roc=[]
def predictions(model):
    pred=model.predict(X_test)
    acc=accuracy_score(pred,y_test)
    a.append(acc)
    pre=precision_score(pred,y_test,zero_division=1)
    p.append(pre)
    recall=recall_score(pred,y_test,zero_division=1)
    r.append(recall)
    f1=f1_score(pred,y_test)
    f.append(f1)
```

```python
predictions(log)
```

```python
predictions(svc)
```

```python
predictions(dtc)
```

```python
predictions(rfc)
```

```python
predictions(knn)
```

```python
def conf_mat(model):
    print(f"Confusion Matrix of {model}")
    pred=model.predict(X_test)
    cm=confusion_matrix(y_test,pred)
    disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
    disp.plot()
    plt.show()
```

```python
conf_mat(log)
conf_mat(svc)
conf_mat(dtc)
conf_mat(rfc)
conf_mat(knn)
```

# Comparing the different Models

```python
model_list=['LogisticRegression','SVC','DecisionTreeClassifier','RandomForestClassifier','KNeighborsClassifier']
comparison=pd.DataFrame({'Model':model_list,
              'accuracy':a,
             'precision':p,
             'recall':r,
             'f1_score':f,
             })
comparison
```

# Plotting Accuracy,Precision,Recall,F1_score

```python
# Set figure size
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(25, 20))

# Plotting accuracy
sns.barplot(x='Model', y='accuracy', data=comparison, ax=axes[0,0], color='blue')
axes[0,0].set_title('Accuracy')
axes[0,0].set_xlabel('Model')
axes[0,0].set_ylabel('Score')
axes[0,0].tick_params(axis='x', rotation=45)
axes[0,0].set_ylim(0.998, 1.00)

# Plotting F1 score
sns.barplot(x='Model', y='f1_score', data=comparison, ax=axes[0,1], color='orange')
axes[0,1].set_title('F1 Score')
axes[0,1].set_xlabel('Model')
axes[0,1].set_ylabel('Score')
axes[0,1].tick_params(axis='x', rotation=45)

# Plotting precision
sns.barplot(x='Model', y='precision', data=comparison, ax=axes[1, 0], color='green')
axes[1, 0].set_title('Precision')
axes[1, 0].set_xlabel('Model')
axes[1, 0].set_ylabel('Score')
axes[1, 0].tick_params(axis='x', rotation=45)
# Set y-axis range for precision plot
axes[1, 0].set_ylim(0.80, 1.00)

# Plotting recall
sns.barplot(x='Model', y='recall', data=comparison, ax=axes[1, 1], color='purple')
axes[1, 1].set_title('Recall')
axes[1, 1].set_xlabel('Model')
axes[1, 1].set_ylabel('Score')
axes[1, 1].tick_params(axis='x', rotation=45)
# Set y-axis range for recall plot
axes[1, 1].set_ylim(0.60, 1.00)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
